```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
le=LabelEncoder()
```

```python
df=pd.read_csv("customer.csv")
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["age"]=pd.to_numeric(df["age"],errors="coerce")
df["age"]=df["age"].fillna(df["age"].mean())
df["gender"]=df["gender"].fillna(df["gender"].mode()[0])
df["city"]=df["city"].fillna("unknown")
```

```python
df["city"]=df["city"].str.strip()
df["city"]=df["city"].str.lower()
```

```python
df.isnull().sum()
```

```python
df["income"]=df["income"].fillna(df["income"].mean())
```

```python
df.isnull().sum()
```

```python
df.duplicated()
```

```python
df[df.duplicated()]
```

```python
df.drop_duplicates()
```

```python
df.info()
```

```python
df=pd.get_dummies(df,columns=["gender","city"])
df
```

```python
df
```

```python
df["age"]=df["age"].astype(int)
```

```python
df
```

```python
x= df.drop("purchased",axis=1)
```

```python
y= df["purchased"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
lo=LogisticRegression(max_iter=1000)
de=DecisionTreeClassifier()
ra=RandomForestClassifier()
kn=KNeighborsClassifier()
```

```python
lo.fit(x_train,y_train)
de.fit(x_train,y_train)
ra.fit(x_train,y_train)
kn.fit(x_train,y_train)
```

```python
lo_pred=lo.predict(x_test)
de_pred=de.predict(x_test)
ra_pred=ra.predict(x_test)
kn_pred=kn.predict(x_test)
```

```python
accuracy_lo=accuracy_score(y_test,lo_pred)
accuracy_de=accuracy_score(y_test,de_pred)
accuracy_ra=accuracy_score(y_test,ra_pred)
accuracy_kn=accuracy_score(y_test,kn_pred)

print("accuracy_lo:", accuracy_lo)
print("accuracy_de", accuracy_de)
print("accuracy_ra:", accuracy_ra)
print("accuracy_kn:", accuracy_kn)
```

```python
new_data=pd.DataFrame([[30,30000,2,5,0,1,0,0,0,0,0,1,0,0]],columns=x.columns)
pred_lo_new=lo.predict_proba(new_data)
pred_de_new=de.predict(new_data)
pred_ra_new=ra.predict(new_data)
pred_kn_new=kn.predict(new_data)

print(pred_lo_new)
print(pred_de_new)
pred_ra_new
pred_kn_new
```
